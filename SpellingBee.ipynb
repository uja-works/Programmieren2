# Funktionale Variante

```python
from functools import reduce

fl = open("Daten/dictionary.txt")
words = fl.read().split('\n')
fl.close()

character_list = "tyiemnf"
center = character_list[0]
char_set = set(character_list)


def score(x):
    val = 1 if len(x) == 4 else len(x)
    # Bonus points for pangrams
    if len(set(x)) == len(char_set):
        val += 7
    return val

solutions = list(filter(lambda x: len(x)>=4 
                  and center in set(x) 
                  and set(x).issubset(char_set), words))
pangrams = list(filter(lambda x: len(set(x)) == len(char_set), solutions))
scored_solutions = list(map(lambda x: (x, score(x)), solutions))
sorted_solutions = list(sorted(scored_solutions, key=lambda x: x[1], reverse=True))
#total_score = sum(n for _, n in sorted_solutions)
total_score = reduce(lambda x, y: x + y[1], scored_solutions, 0)

print(pangrams)
print(sorted_solutions)
print(f"Maximale mögliche Punktzahl = {total_score}")
```
