## Aufgabe 1
Berechnen Sie mit Hilfe von Sympy das bestimmte Integral
$$ I = \int_a^b \frac{1+x}{1-2x^2} \, dx. $$

**Lösung:** Importieren von `sympy`, Definition von $x, a, b$ als Symbole und Verwenden der Funktion `integrate`.

```python
from sympy import *

x, a, b = symbols("x a b")
integrate( (1 + x) / (1 - 2*x**2), (x, a, b) )
```

## Aufgabe 2
Bestimmen Sie das berühmte Gaußsche Integral
$$ \int_{-\infty}^\infty e^{-x^2} \,dx. $$

**Lösung:** In Sympy ist `oo` das Symbol für "unendlich". Das obige Integral muss $\sqrt{\pi}$ ergeben.

```python
integrate( exp(-x*x), (x, -oo, oo))
```

## Aufgabe 3
Bestimmen Sie mit Sympy die Summen
\begin{align}
 S_1 &= \sum_{k=1}^n k, \\      % Zeilenumbruch: \\, Tabulator: &
 S_2 &= \sum_{k=1}^n k^2, \\
 S_3 &= \sum_{k=3}^\infty q^3 \qquad \text{für } |q|<1.
\end{align}

**Lösung:**

```python
k, n = symbols("k n")
S_1 = Sum(k, (k, 1, n)) 
S_1.doit().simplify()
```

```python
# Alternative: summation statt Sum...doit()
S_2 = summation(k**2, (k, 1, n)) 
S_2.simplify().factor()
```
